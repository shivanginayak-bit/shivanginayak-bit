```python
x = lambda a,b : a+b
x(1,5)
```

```python
# lambda function - Row wise operation
def revised_profit(revenue, budget):
  if revenue > 0:
    new_profit = revenue - budget
  else:
    new_profit = np.nan
    
  return new_profit
```

```python
df['new_profit'] = df.apply(lambda x: revised_profit(x['revenue'], x['budget']),axis=1)
```

def grad_descent
deciding when to stop the algorithm


```python
def grad_descent(derivative,x_prev,learning_rate):
  epsilon=0.001
  grad_iter=1
  x_next = x_prev - learning_rate*derivative(x_prev)
  sequence = np.array([])
  sequence=np.append(sequence,x_next)
  while abs(x_next-x_prev) > epsilon:
   x_prev=x_next
   x_next = x_prev - learning_rate*derivative(x_prev)
   sequence=np.append(sequence,x_next)
   grad_iter = grad_iter + 1
   return x_next,sequence,grad_iter
```
