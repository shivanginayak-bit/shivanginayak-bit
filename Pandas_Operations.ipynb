def grad_descent
deciding when to stop the algorithm


````python
def grad_descent(derivative,x_prev,learning_rate):
  epsilon=0.001
  grad_iter=1
  x_next = x_prev - learning_rate*derivative(x_prev)
  sequence = np.array([])
  sequence=np.append(sequence,x_next)
  while abs(x_next-x_prev) > epsilon:
   x_prev=x_next
   x_next = x_prev - learning_rate*derivative(x_prev)
   sequence=np.append(sequence,x_next)
   grad_iter = grad_iter + 1
   return x_next,sequence,grad_iter
````

````python
np.random.seed(123)
MARKER_SIZE = 100
DATA_COLOR = 'black'
ERROR_COLOR = 'darkred'
POLYNOMIAL_FIT_COLORS = ['orange', 'royalblue', 'darkgreen']
LEGEND_FONTSIZE = 14
TITLE_FONTISIZE = 16
N_OBSERVATIONS = 10
NOISE_STD = 1.

x = 2 * (np.random.rand(N_OBSERVATIONS) - .5)
x_grid = np.linspace(-1, 1, 100)


def f(x):
    """Base function"""
    return np.sin(x * np.pi)


def sample_fx_data(shape, noise_std=NOISE_STD):
    return f(x) + np.random.randn(*shape) * noise_std


def plot_fx_data(y=None):
    """Plot f(x) and noisy samples"""
    y = y if y is not None else sample_fx_data(x.shape)
    fig, axs = plt.subplots(figsize=(6, 6))
    plt.plot(x_grid, f(x_grid), color=DATA_COLOR, label='f(x)')
    plt.scatter(x, y, s=MARKER_SIZE, edgecolor=DATA_COLOR, facecolors='none', label='y')


# Plot the data
y = sample_fx_data(x.shape)
plot_fx_data(y)
plt.legend(fontsize=14)
plt.title(f'$f(x) = sin(\pi x)$ and some observations, $y$', fontsize=16)
plt.xlim([-1, 1])
plt.ylim([-2, 2])
````

````python
def factorial(n):
  f=1
  for i in range(1,n+1):
   f=f*i
  return f
````

````python
def P(n,r):
  return int(factorial(n)/factorial(n-r))
````

````python
def C(n, r):
  return int(factorial(n)/(factorial(r) * factorial(n-r)))
````
